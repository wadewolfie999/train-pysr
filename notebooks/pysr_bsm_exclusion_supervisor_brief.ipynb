## Minimal Research Setup

The operational positive class used by the saved scripts is numeric label `1`, inferred from the binary labels and still marked pending review. The physical meaning of target label `1` is therefore not asserted here. ROC-AUC is computed from continuous saved scores, not hard labels.


```python
from pathlib import Path
import json

candidates = [
    Path("notebooks/report_data/pysr_bsm_exclusion_supervisor_brief_data.json"),
    Path("report_data/pysr_bsm_exclusion_supervisor_brief_data.json"),
    Path.cwd().parent / "notebooks/report_data/pysr_bsm_exclusion_supervisor_brief_data.json",
]
data_path = next(path for path in candidates if path.exists())
data = json.loads(data_path.read_text(encoding="utf-8"))
assert data["candidate"]["fixed_split_roc_auc"] == data["candidate"]["auc_from_saved_scores"]
```

## Result Snapshot

| Result | Method | Validation | ROC-AUC | Note |
| --- | --- | --- | ---: | --- |
| Initial raw two-feature baseline | Random forest | Fixed seed-42 stratified test split | 0.918117 | standard ML baseline; provisional |

| Repeated-split robustness | HistGradientBoostingClassifier | 10 stratified repeated splits | 0.969264 | mean; min 0.964378 |
| Cross-validation robustness | HistGradientBoostingClassifier | 5-fold stratified CV | 0.967593 | mean; below 0.97 |

The table keeps the emphasis on the decision-relevant comparison: the engineered fixed-split candidate crosses the target, while the repeated-split and cross-validation summaries do not. The raw two-feature value is included only as the initial baseline context, not as the strongest later raw-mass result.


```python
import matplotlib.pyplot as plt

roc = data["candidate"]["roc_curve"]
auc_value = data["candidate"]["fixed_split_roc_auc"]
fig, ax = plt.subplots(figsize=(5.8, 4.0))
ax.plot(roc["fpr"], roc["tpr"], color="#0f766e", linewidth=2.2, label=f"Engineered HGB, AUC = {auc_value:.4f}")
ax.plot([0, 1], [0, 1], color="#8a8f98", linestyle="--", linewidth=1.0, label="Chance")
ax.set_xlabel("False positive rate")
ax.set_ylabel("True positive rate")
ax.set_title("Fixed seed-42 ROC from saved continuous scores")
ax.set_xlim(0, 1)
ax.set_ylim(0, 1.02)
ax.grid(True, alpha=0.25)
ax.legend(loc="lower right", frameon=False)
fig.tight_layout()
```

The result is useful because it indicates that approved mass-derived structure can substantially improve over the initial two-feature raw-mass baseline.
